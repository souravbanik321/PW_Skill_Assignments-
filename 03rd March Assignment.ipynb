## Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Some of the popular plots that we can plot using Seaborn are:

- `Scatter plot`: A scatter plot shows the relationship between two variables by plotting the points on a two-dimensional plane. Seaborn's scatter plot adds additional functionality to Matplotlib's scatter plot, such as the ability to plot multiple groups with different colors and markers, and the ability to fit a linear regression line to the data.

- `Line plot`: A line plot displays the trend of a variable over time or across different categories. Seaborn's line plot can plot multiple lines on the same plot, with each line representing a different category or variable.

- `Bar plot`: A bar plot shows the comparison between different categories or groups using vertical or horizontal bars. Seaborn's bar plot can display the mean or median of the data, and can also display the standard deviation or confidence interval.

- `Histogram`: A histogram is a graphical representation of the distribution of a dataset. Seaborn's histogram adds additional functionality to Matplotlib's histogram, such as the ability to plot multiple histograms on the same plot, and the ability to fit a kernel density estimate (KDE) to the data.

- `Heatmap`: A heatmap is a graphical representation of a matrix, where each cell is colored according to its value. Seaborn's heatmap can display the correlation between variables, or the distribution of values across different categories.

Each of these plots can be useful for different types of data and different types of analysis. For example, a scatter plot can be useful for visualizing the relationship between two continuous variables, while a bar plot can be useful for comparing the means of different groups. A heatmap can be useful for identifying patterns or correlations in large datasets.

```python
import seaborn as sns

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot using "timepoint" as x-axis and "signal" as y-axis, grouped by "event" and "region"
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

## Q3. Load the "titanic" dataset using the load_dataset function of seaborn. 
Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot
sns.boxplot(x="pclass", y="age", data=titanic_data)

# Create the second box plot
sns.boxplot(x="pclass", y="fare", data=titanic_data)

```

## Q4.Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create the histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut")
```

## Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create the pair plot with hue
sns.pairplot(data=iris_data, hue="species")
```

## Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Reshape the data to wide format
flights_pivot = flights_data.pivot(index="year", columns="month", values="passengers")

# Create the heatmap
sns.heatmap(data=flights_pivot, cmap="Blues", annot=True, fmt="d")
```
