Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is classified as a micro framework because it does not require particular tools or libraries, unlike full-stack frameworks like Django. Flask is lightweight and flexible, making it an ideal choice for developing small to medium-sized web applications.

`Advantages of Flask Framework include:`

1. Lightweight: Flask is a lightweight framework that does not require any specific tools or libraries, making it easy to install and use.

2. Flexible: Flask is very flexible, allowing developers to customize the framework to suit their needs. It does not enforce any particular programming paradigm, so developers can use their preferred approaches.

3. Modular: Flask is modular, allowing developers to use only the components they need. This helps to keep the application size small and reduces dependencies.

4. Easy to learn: Flask is easy to learn and use, making it a good choice for beginners who are just starting with web development.

5. Extensible: Flask is extensible, allowing developers to add third-party extensions to enhance functionality, such as authentication and database integration.

6. Testable: Flask is testable, making it easy to write automated tests for the application.

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is a way to map URLs or routes to functions that handle them. It is a fundamental concept in Flask and allows developers to define how different parts of the application respond to incoming requests.

We use app routes in Flask to define the URLs or endpoints that the application responds to. Each route is associated with a function that handles the request and returns a response. When a request is made to a specific URL, Flask uses the app route to determine which function should handle the request.

App routes are useful in Flask for several reasons:

1. URL mapping: App routes allow developers to map specific URLs to functions that handle them. This makes it easy to define the structure of the application and to organize code effectively.

2. Handling requests: App routes allow developers to define how the application should respond to incoming requests. This can include returning specific data or rendering a template.

3. Dynamic routing: App routes can also be used to create dynamic URLs that include variables. This makes it easy to handle requests for resources that vary based on user input, such as search queries or user profiles.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL to a specific function dynamically, based on the name of the function and the arguments provided to it. This makes it easy to create links between different parts of a Flask application.

Here is an example code that demonstrates the working of the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # create a link to the 'about' page
    about_url = url_for('about')
    # return the link as HTML
    return f"Click <a href='{about_url}'>here</a> to go to the 'about' page."

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run()
```

In this example, we define two routes: / and /about. The index() function generates a link to the about() function using the url_for() function, and returns it as HTML. When the user clicks on the link, they will be taken to the about() page.

When we run the application and navigate to http://localhost:5000/, we will see a link that says "Click here to go to the 'about' page." If we click on this link, we will be taken to the about() page, where we will see the text "This is the about page." displayed on the screen.

The url_for() function is a powerful feature in Flask that makes it easy to create dynamic links between different parts of a web application. By using this function, we can avoid hard-coding URLs in our code and make our application more flexible and maintainable.
