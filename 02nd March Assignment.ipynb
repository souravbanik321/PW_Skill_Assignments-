## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib. 

Matplotlib is a popular Python library used for data visualization. It is a powerful tool for creating a wide range of static, interactive, and animated visualizations in Python. Matplotlib is widely used in scientific computing, data analysis, and machine learning applications.

Matplotlib is used for creating high-quality plots, charts, histograms, scatterplots, and other visualizations to help users understand complex data patterns and trends. It provides a range of customizable features for creating plots and charts, including the ability to adjust the color, style, and size of lines, markers, and fonts.

The Pyplot module of Matplotlib is a sublibrary that provides a convenient interface for creating a variety of plots with a few lines of code. Here are five plots that can be created using the Pyplot module of Matplotlib:

- `Line Plot` - used to represent the relationship between two variables by drawing a line between the data points on a Cartesian plane.

- `Scatter Plot` - used to represent the relationship between two variables by placing individual data points on a Cartesian plane.

- `Bar Plot` - used to compare the values of different categories by representing them as rectangular bars.

- `Histogram `- used to represent the distribution of a continuous variable by dividing it into intervals and counting the number of observations in each interval.

- `Pie Chart` - used to represent the relative size of different categories as slices of a circular pie.





## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
#import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of plot that displays the relationship between two variables by plotting individual data points on a two-dimensional Cartesian plane. Each point in a scatter plot represents the value of the two variables for a single observation in the dataset. Scatter plots are useful for identifying patterns and trends in the data, as well as for detecting outliers and other unusual observations.

Here's the code to generate a scatter plot using the data provided:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

The `scatter()` function from the Pyplot module of Matplotlib is used to create the scatter plot. The `title()`, `xlabel()`, and `ylabel()` functions are used to add a title and axis labels to the plot. The `show()` function is called to display the plot. The resulting scatter plot should show a cloud of points scattered around the center of the plot, indicating a random distribution of values for both variables.

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python
#import numpy as np
#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The `subplot()` function in Matplotlib is used to create multiple plots within a single figure. This function creates a grid of subplots where each subplot can be customized to display a different plot. The `subplot()` function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

Here's the code to create four line plots using the `subplot()` function:

```python
import numpy as np
import matplotlib.pyplot as plt

# create the data for each line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a figure with four subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# plot each line on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# add a title for the entire figure
fig.suptitle('Four Line Plots')

# adjust the spacing between subplots
plt.subplots_adjust(hspace=0.4)

# display the plot
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
#import numpy as np
#company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is a graphical representation of data that uses rectangular bars to show the magnitude of the values being plotted. It is typically used to compare values across different categories.

In the given data, the company array contains the names of four technology companies, and the profit array contains the corresponding profits of each company in millions of dollars.

To create a vertical bar plot using this data, we can use the bar function from the matplotlib.pyplot module:

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Profits of Technology Companies")
plt.xlabel("Company")
plt.ylabel("Profit (in millions of dollars)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company, profit)
plt.title("Profits of Technology Companies")
plt.xlabel("Profit (in millions of dollars)")
plt.ylabel("Company")
plt.show()
```

A box plot is a graphical representation of the distribution of a set of data. It displays the median, quartiles, and outliers of the data in a box-and-whisker format. It is used to identify the spread and skewness of the data, as well as any potential outliers.

In the given data, box1 and box2 are two NumPy arrays that contain 200 randomly generated values each, with mean and standard deviation parameters.

To create a box plot using this data, we can use the boxplot function from the matplotlib.pyplot module:

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title("Box Plot of Box 1 and Box 2")
plt.xlabel("Box")
plt.ylabel("Values")
plt.show()
```
